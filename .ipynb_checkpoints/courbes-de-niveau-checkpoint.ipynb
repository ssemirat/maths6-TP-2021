## Descente de gradient en 2 variables

### Import des librairies

```python
from matplotlib import pyplot as plt
import numpy as np
```

> Fonction python permettant de dessiner les courbes de niveaux d'une fonction de 2 variables

```python
def courbes_de_niveaux(fct, x1_min=-5, x2_min=-5, x1_max=5, x2_max=5,k=40):
    x1_values = np.linspace(x1_min, x1_max,100)
    x2_values = np.linspace(x2_min, x2_max,100)
    fct_values = np.array([[fct(x1,x2) for x1 in x1_values] for x2 in x2_values])    
    plt.contour(x1_values, x2_values, fct_values, k)
    plt.xlim((x1_min, x1_max))
    plt.ylim((x2_min, x2_max))
    plt.colorbar()
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
def H1(x1,x2):
    return sigmoid(1-3*x1+3*x2)
def H2(x1,x2):
    return sigmoid(-1-3*x1+3*x2)

plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.title("$\sigma(1-3x_1+3x_2)$")
courbes_de_niveaux(H1,0,0,1,1)
plt.subplot(1,2,2)
plt.title("$\sigma(-1-3x_1+3x_2)$")
courbes_de_niveaux(H2,0,0,1,1)

```

```python
def f1(x1,x2):
    return 2*x1+2*x2
def f2(x1,x2):
    return 2*x1-2*x2
def f3(x1,x2):
    return -2*x1+2*x2
def f4(x1,x2):
    return -2*x1-2*x2
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.title("$a=2$, $b=2$")
courbes_de_niveaux(f1,0,0,1,1)
plt.subplot(2,2,2)
plt.title("$a=2$, $b=-2$")
courbes_de_niveaux(f2,0,0,1,1)
plt.subplot(2,2,3)
plt.title("$a=-2$, $b=2$")
courbes_de_niveaux(f3,0,0,1,1)
plt.subplot(2,2,4)
plt.title("$a=-2$, $b=-2$")
courbes_de_niveaux(f4,0,0,1,1)
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
def f1(x1,x2):
    return sigmoid(3*x1+3*x2)
def f2(x1,x2):
    return sigmoid(3*x1-3*x2)
def f3(x1,x2):
    return sigmoid(-3*x1+3*x2)
def f4(x1,x2):
    return sigmoid(-3*x1-3*x2+1)
plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
plt.title("$a=3$, $b=3$")
courbes_de_niveaux(f1,0,0,1,1)
plt.subplot(2,2,2)
plt.title("$a=3$, $b=-3$")
courbes_de_niveaux(f2,0,0,1,1)
plt.subplot(2,2,3)
plt.title("$a=-3$, $b=3$")
courbes_de_niveaux(f3,0,0,1,1)
plt.subplot(2,2,4)
plt.title("$a=-3$, $b=-3$")
courbes_de_niveaux(f4,0,0,1,1)
```
