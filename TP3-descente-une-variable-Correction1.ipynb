```python
# Chargement des librairies Python

# Mathématiques 
import numpy as np

# Graphique
from matplotlib import pyplot as plt
```

## Problème

On cherche le minimum sur $\mathbb{R}$ de la fonction $f(x)=x^2-6x+1$.

**Exercice 1.**

Résoudre le problème par le calcul.

***Correction***

Je calcule la dérivée : $f'(x)=2x-6$.
Ainsi $f'(x)>0$ pour $x>3$, $f'(x)<0$ pour $x<3$.
Donc $f$ atteint un minimum en $x=3$.

### Fonction à optimiser et sa dérivée

```python
# Fonction à optimiser
def f(x):           
  return x**2-6*x+1

# fonction dérivée
def df(x):          
  return 2*x-6
```

### Représentation graphique de $f$

```python
plt.axis([-1,6,-10,5]) #xmin,xmax,ymin,ymax
listx= np.linspace(-2,6,100) # Génère 100 nombres entre -2 et 6 et régulièrement espacés
plt.plot(listx,[f(x) for x in listx]) # Représentation de la fonction
```

### Algorithme de descente de gradient

```python
def descente(fonctderivee,x0,tau,tolerance,nbiterations):
    x=x0
    L=[x]
    for i in range(nbiterations): #i parcourt de 0 à nbiterations
        d = fonctderivee(x)
        if abs(d)< tolerance: # abs renvoie la valeur absolue
            print('L\'algorithme a convergé en',i,'itérations.\nSolution atteinte : x=',x,'\nDérivée en x :',d)
            return L
        x=x-tau*d
        L.append(x) # on enregistre chaque valeur dans le tableau L pour la représentation graphique ci-dessous
    print('L\'algorithme n\'a pas convergé.\nSolution atteinte: x=',x,'\nDérivée en x :',d)
    return L
```

**Exercice 2.** ***Correction***

Décrire et donner le rôle de chacun des arguments de `descente(fonctderivee,x0,tau,tolerance,nbiterations)` ci-dessus :
- fonctderivee : *la dérivée de la fonction dont le minimum*
- x0 : *le point à partir duquel on initialise l'algorithme*
- tau : *le pas de l'algorithme*
- tolerance : *le paramètre en deça duquel on estime être assez prêt de 0 pour arrêter l'algorithme*
- nbiterations : *le nombre maximum de calcul de x(i+1) en fonction de x(i) : au-delà de ce nombre on arrête l'algorithme même si on n'a pas trouvé une dérivée plus petite que tolerance.*

## Recherche de la solution

```python
Xi=descente(df,0.5,0.8,1e-5,200)
# Ici : 
# fonctionderivee=2x-6 (la dérivée de la fonction à minimiser)
# x0=0.5 : on part de 0.5
# tau=0.8 : le pas
# tolerance=1e-5 : on cherche f'(x) tel que |f'(x)|<0.00001
# nbiteration=200 : on arrête au bout de 200 iterations même si on n'a pas trouvé la solution
```

### Liste des points atteints par l'algorithme

```python
print(Xi)
```

### Représentation graphique de l'algorithme

```python
plt.axis([-1,6,-10,5]) #xmin,xmax,ymin,ymax
listx= np.linspace(-2,6,100) # Génère 100 nombres régulièrement espacés de -2 à 6
plt.plot(listx,[f(x) for x in listx]) # Représentation de la fonction
plt.plot(Xi,[f(x) for x in Xi],'o--') # Représentation des segments (x_i,f(x_i))--(x_(i+1),f(x_(i+1)))
plt.title('Descente de gradient en 1 variable')
plt.show
```

***Correction***

```python
Xi=descente(df,0.5,0.3,1e-5,200)
print(Xi)
plt.axis([-1,6,-10,5]) #xmin,xmax,ymin,ymax
listx= np.linspace(-2,6,100) # Génère 100 nombres régulièrement espacés de -2 à 6
plt.plot(listx,[f(x) for x in listx]) # Représentation de la fonction
plt.plot(Xi,[f(x) for x in Xi],'o--') # Représentation des segments (x_i,f(x_i))--(x_(i+1),f(x_(i+1)))
plt.title('Descente de gradient en 1 variable')
plt.show
```

2. Pourquoi $\tau=0.5$ est toujours optimal quel que soit $x_0$ ?

A partir de $x_0$, la valeur de $x_1$ est : $x_0-\tau \times f'(x_0)=x_0-\tau\times (2x_0-6)$

Si $\tau=0.5$, on obtient : $x_1=x_0-0.5(2x_0-6)=x_0-x_0+3=3$ quel que soit la valeur de $x_0$.
